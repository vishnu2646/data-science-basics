## Seaborn
 - A advanced library, built on top of Matplotlib, used for data visualization.
 - Also called as EDA (Explotratory Data Analysis)
       - create some charts and graphs to get usefull insights about the data.
   #### Some Types of Analysis:
   
 - univariate Analysis
       - single column | variable
 - bivariate Analysis
       - Two column | variables
 - multivariate Analysis
       - More than two columns

```python
import matplotlib.pyplot as plt 
import seaborn as sns
import warnings
import pandas as pd

warnings.filterwarnings('ignore')
```

```python
sns.get_dataset_names()
```

```python
df = pd.read_csv('./iris.csv')
```

```python
df
```

```python
## HistPlot
```

```python
sns.histplot(data=df, x='SepalLengthCm', kde=True) # kde = kernal density curve
```

```python
sns.histplot(data=df, x='SepalLengthCm', kde=True, hue='Species')
```

## Distplot

```python
sns.distplot(df.PetalLengthCm, hist=False, color="blue") # will nor Accept any args ather than the hist and color
```

## CoutPlot

```python
sns.countplot(data=df, x='Species', palette=['#09b341', '#435be6', '#12c79a'])
```

## boxplot

```python
sns.boxplot(data=df, x='SepalWidthCm', color="hotpink")
```

```python
sns.boxplot(data=df, y='SepalWidthCm', x='Species', palette=['#3af525', '#30dbd8', '#ec24f0'])
```

```python
sns.violinplot(data=df, x='SepalWidthCm',color='#30dbd8')
```

```python
sns.violinplot(data=df, y='SepalWidthCm', x='Species', palette=['#3af525', '#30dbd8', '#ec24f0'])
```

## barplot

```python
sns.barplot(data=df, x='Species', y='PetalLengthCm')
```
