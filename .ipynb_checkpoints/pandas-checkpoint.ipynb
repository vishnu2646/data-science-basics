```python
# deleting 
# adding the inplace=True completely removes the data from the series.
# s3.drop('Boolean', inplace=True)
s3.drop('Boolean')
```

```python
# deleting multiple values
# this is only possible in the custom value of idnex
s3.drop(['integer', 'Float'])
```

## DataFrame
 - A Pandas DataFrame is a 2 Dimenasional data structure, like a 2 dimensional array, or a table with rows and columns
 - Dataframe can have any number of dimensions

```python
df = pd.DataFrame()
df
```

```python
type(df)
```

```python
lst1 = [['abc',1], ['def',2],['ghi',3], ['jkl'],['mnp',5]]
lst1
```

```python
df1 = pd.DataFrame(lst1, columns=['A', 'B'], index=['one', 'two', 'three','four', 'five'])
df1 # if failed to give value then by default it takes to NaN or None
```

```python
df1.rename({'A': 'a', 'B':'b'}, axis=1, inplace=True)
df1
```

```python
d1 = { 'name': ['karthik', 'shree', 'shankar', 'monica', 'kavipriya'], 
       'marks':[100,90, 95, 85, 92],
       'age': [24,26, 23, 27, 22]
     }
df3 = pd.DataFrame(d1, index=[1,2,3,4,5])
df3
```

```python
df3.rename({'name':'Name', 'marks': 'Marks', 'age': 'Age'}, axis=1, inplace=True)
df3
```

```python
df3['City']="Chennai" # adding new column with single value
```

```python
df3['Gender']= ['Male', 'Female', 'Male', 'Female', 'Female'] # adding column with multiple values equal to entries
df3
```

```python
df3.drop('City', axis=1, inplace=True)
df3
```

```python
df3.drop(['Marks', 'Age'], axis=1)
```

```python
df3.drop(5, axis=0)
```

## Accesing datafram
 - iloc (using index numbers in dataframe)
 - loc (using index, column name)

```python
df3
```

```python
# iloc -> indexing
df3.iloc[0,1]
```

```python
df3.iloc[4, 2]
```

```python
# iloc -> slicing
df3.iloc[3:,1:3]
```

```python
df3.iloc[0,0] = "siva" # updating single value
df3
```

```python
df3.iloc[0:2,1] = [95, 91] # updating multi values
df3
```

```python
# loc -> indexing
df3.loc[2, 'Name']
```

```python
df3.loc[3:,:'Marks']
```

```python
# updating the data in dataframe.

df3.loc[4:, 'Age'] = [25, 24]
df3
```

```python
df3.loc[4:, 'Marks': 'Age'] = [[95, 26], [90, 25]] # the input taken with row axis.
df3
```

## Imputing -> Add the mean values in palces of null values

```python
np.mean(data.Age) # mean value of the Age column
```

```python
data.Cabin.fillna(method='ffill')  # Add inplace to True in command to change the oiginal data
```

```python
data.Cabin.fillna(method='bfill')  # Add inplace to True in command to change the oiginal data
```

```python
data.Age.fillna(np.mean(data.Age)) # Add inplace to True in command to change the oiginal data
```

```python
data.loc[(data.Age).isnull() == True] = np.mean(data.Age)
```

## Grouping the data
  - used to compare two columns(1 catrgorial, 1 numerical), should use aggregate function for numerical column
  - single column is also possible

  - Aggregate functions - sum, count, mean, max, min

```python
data.groupby('Sex').count()
```

```python
data.groupby('Sex')['Fare'].max()
```

```python
data.groupby('Survived')['Fare'].min()
```

```python
data.groupby('Survived')['Fare'].max()
```

```python
data.groupby('Pclass')['Fare'].max()
```

```python
data.groupby('Pclass')['Fare'].mean()
```
